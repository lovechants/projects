### Background 
This data presents the top ten leading causes of death in the United States beginning in 1999 from the [NCHS (National Center for Health Statistics)](https://www.cdc.gov/nchs/index.htm). The data is taken from all 50 states and includes the District of Columbia. Causes of death are taken from and classified by the International Classification of Diseases, and cause of death statistics are based on the underlying causes.

#### Data Preface
As reported by [U.S. Department of Health & Human Services](https://catalog.data.gov/dataset/nchs-leading-causes-of-death-united-states) the data has been normalized under these parameters:
> __Age Adjusted__ : Age-adjusted death rates (per 100,000 population) are based on the 2000 U.S. standard population. [Definition](https://www.health.pa.gov/topics/HealthStatistics/Statistical-Resources/UnderstandingHealthStats/Pages/Age-Adjusted-Rates.aspx)

> __Population Statistics__ : Populations used for computing death rates after 2010 are postcensal estimates based on the 2010 census, estimated as of July 1, 2010

> Further reference can be found through [NCHS](https://www.cdc.gov/nchs/data_access/vitalstatsonline.htm) and [Murphy SL, Xu JQ, Kochanek KD, Curtin SC, and Arias E. Deaths: Final data for 2015](https://www.cdc.gov/nchs/data/nvsr/nvsr66/nvsr66_06.pdf)

### Data Loading and Cleaning
To start our analysis we begin by importing libaries that will be useful to preprocessing and any data cleansing that is needed.
We will also read in our data and store it as a data frame and look at the intial entries.
These are the steps we will take before interpret and evaluate our data.
1. Convert our data
2. Check missing values
3. Check for outliers
4. Feature selection
5. Split 
6. Evaluate 

```python
# Import libraries for process 
import pandas as pd
import numpy as np
# Reading in the data 
leading_causes_of_death = pd.read_csv("data/nchs.csv") 
leading_causes_of_death.head()
```

#### 1. Convert our Data

```python
leading_causes_of_death.info() 
```

The column names "113 Cause Name" and "Cause Name" sound similar, however, column name "113 Cause Name" is icd-10 standard code, and the full [table](https://www-doh.state.nj.us/doh-shad/query/NCHS113.html) can be found on the WHO website. Since we can use the codes as a standard to evaluate and we already have the name we will split the cause name from the cause code and rename the columns.

```python
# Create a function to split the string 
def split_data(column):
    return pd.Series(column.split("("))
# Apply it to our data
leading_causes_of_death[["Underlying Cause", "Extra", "113 Code"]] = leading_causes_of_death["113 Cause Name"].apply(split_data)
leading_causes_of_death
```

Now that we've split the "113 Cause Name" column we are left with three extra columns, to further clean this up we can now drop the original "113 Cause Name Column" and we can see that some cause have additional underlying causes, and our codes didn't all split to a seperate column. After dropping our column we can reformat our data to clean it up a little bit removing the ")" since our delimiter from our function was "(". 

```python
# removing the ")" from our split data
leading_causes_of_death['113 Code'] = leading_causes_of_death['113 Code'].str.replace(')', '')
leading_causes_of_death['Extra'] = leading_causes_of_death['Extra'].str.replace(')', '')

# drop 113 Cause Name column
leading_causes_of_death.drop('113 Cause Name', axis=1, inplace=True)
leading_causes_of_death
```

To finish cleaning our dataset we need to deal with our newly created "Extra" and "113 Code" columns.

```python
set(leading_causes_of_death.Extra)
```

Looking at terms in the "Extra" column we firstly need to move the codes over to the "113 code" column. After we can drop the "Extra" column as there are only two different causes that provide us with extra descriptors, in which the data already describes these cases.

```python
# Copy data from our split from our extra to our 113 code colum

# Grabbing our indices where the 113 code is NaN 
empty_codes = leading_causes_of_death['113 Code'].isnull()

# Setting copying the 113 code into the correct column 
leading_causes_of_death.loc[empty_codes, '113 Code']  = leading_causes_of_death.loc[empty_codes, 'Extra']

# Drop our extra column 
leading_causes_of_death.drop('Extra', axis=1, inplace=True)
leading_causes_of_death
```

#### 2. Check missing values


```python
# Grabbing the total of all null values
total_null = leading_causes_of_death.isnull().sum()
print(total_null)

# Examining the data where the code is null
emptyCodes = leading_causes_of_death.isnull().any(axis=1)
emptyDF = leading_causes_of_death[emptyCodes]
emptyDF
```

Checking our missing data it looks like these are total numbers for "All causes" or a total of all deaths per state and for the United States that year for each year we have in our dataset (1999-2017). To finish cleaning, we will remove these entries from our data frame and store it in a seperate data frame. In addition to that we will remove all total entries where our state is "United States" as that is the country total per each cause; we will store this in the same data frame as "all causes".

```python
all_causes = leading_causes_of_death[leading_causes_of_death['113 Code'].isna()]
all_causes
```

```python
leading_causes_of_death.dropna(subset=['113 Code'], inplace=True)
total_null = leading_causes_of_death.isnull().sum()
total_null
```

```python
totals_df = leading_causes_of_death[leading_causes_of_death['State'] == 'United States'].copy()
totals_df
```

```python
leading_causes_of_death = leading_causes_of_death[leading_causes_of_death['State'] != 'United States']
```

## Visualization and Descriptive Statistics
---
Now that we have cleaned our data and created the necessary dataframes to store correlated data we can be begin to start exploring and visualizing our data.



```python
# visualizations to make: Per year, death rate per state
```

```python
# Import libraries for visualizatins
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pie chart for the proportion of deaths by cause in the most recent year

# Filter the data for the most recent year
recent_pie_data = leading_causes_of_death[leading_causes_of_death['Year'] == 2017].groupby('Cause Name')['Deaths'].sum()
past_pie_data = leading_causes_of_death[leading_causes_of_death['Year'] == 1999].groupby('Cause Name')['Deaths'].sum()

# Create a pie chart
plt.figure(1, figsize=(10, 8))
plt.subplot(1,2,1)
recent_year_pie = plt.pie(recent_pie_data, labels=recent_pie_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'),textprops={'fontsize': 7})
plt.title(f'Percentage of Deaths by Cause in 2017')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.plot()

# Create a pie chart
plt.figure(2, figsize=(10, 8))
plt.subplot(1, 2, 2)
plt.pie(past_pie_data, labels=past_pie_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), textprops={'fontsize': 7})
plt.title(f'Percentage of Deaths by Cause in 1999')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.plot()

plt.show()



```

Anaylsis on this 

```python
# Filter and group the data for all causes
trend_per_cause = leading_causes_of_death.groupby(['Year', 'Cause Name'])['Deaths'].sum().unstack()

# Create line graphs for all causes
plt.figure(figsize=(15, 10))
for cause in trend_per_cause.columns:
    plt.plot(trend_per_cause.index, trend_per_cause[cause], label=cause, marker='o')

plt.title('Trend of Deaths for All Causes Over the Years')
plt.ylabel('Number of Deaths')
plt.xlabel('Year')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place the legend outside the plot
plt.grid(True)
plt.show()

```

```python

recent_year_stats = leading_causes_of_death[leading_causes_of_death['Year'] == 2017][['Cause Name', 'Age-adjusted Death Rate']]
past_year_stats = leading_causes_of_death[leading_causes_of_death['Year'] == 1999][['Cause Name', 'Age-adjusted Death Rate']]


plt.figure(figsize=(16, 10))
# Create a box plot
plt.subplot(2, 2, 2)
sns.boxplot(x='Age-adjusted Death Rate', y='Cause Name', data=recent_year_stats, palette='Set2')
plt.title(f'Comparison of Age-Adjusted Death Rates by Cause in 2017')
plt.xlabel('Age-Adjusted Death Rate')
plt.ylabel('Cause of Death')
plt.plot()

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, position 1
sns.histplot(recent_year_stats['Age-adjusted Death Rate'],  color='blue')
plt.title('Histogram of Age-Adjusted Death Rates in 2017')
plt.xlabel('Age-Adjusted Death Rate')
plt.ylabel('Frequency')


plt.subplot(2, 2, 4)
sns.boxplot(x='Age-adjusted Death Rate', y='Cause Name', data=past_year_stats, palette='Set2')
plt.title(f'Comparison of Age-Adjusted Death Rates by Cause in 1999')
plt.xlabel('Age-Adjusted Death Rate')
plt.ylabel('Cause of Death')
plt.plot()

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, position 3
sns.histplot(past_year_stats['Age-adjusted Death Rate'],  color='green')
plt.title('Histogram of Age-Adjusted Death Rates in 1999')
plt.xlabel('Age-Adjusted Death Rate')
plt.ylabel('Frequency')

plt.show()
```
